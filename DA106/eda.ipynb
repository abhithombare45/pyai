```python
import seaborn as sns   #Visualization and plotting purpose 
import matplotlib.pyplot as plt  #plotting purpose 
```

```python
#Load IRIS Dataset
df = sns.load_dataset('iris')
```

```python
df
```

```python
#Histogram
plt.figure(figsize=(10,6))
sns.histplot(df['sepal_length'],bins=30)
plt.title('Histogram of Sepal Lenght')
plt.show()
```

```python
#Density Plot
plt.figure(figsize=(10,6))
sns.kdeplot(df['sepal_length'])
plt.title('Density Plot of Sepal Lenght')
plt.show
```

```python
#Box Plot
plt.figure(figsize=(10,6))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Box Plot of Sepal Length')
plt.show
```

```python
#Correlation Plot
plt.figure(figsize=(10,6))
sns.heatmap(df.iloc[:,0:4].corr(), annot=True, cmap='çoolwarm')
plt.title('Çorrelation Plot of Iris Dataset')
plt.show()
```
